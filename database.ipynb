```python
import mysql.connector
from tqdm import tqdm
conn = mysql.connector.connect(user='root', password='root', host='localhost', database='irproject')

#Creating a cursor object using the cursor() method
cursor = conn.cursor()
not_added = []
count = 0
with open("irdatabase.csv", "r") as fin:
    #line = fin.readline()
    for line in tqdm(fin.readlines()):
        count+=1
        vals = line.split(',')
        word = vals[0]
        pairs = vals[1]
        corpus_count = vals[2]
        docs_count = vals[3]
        dataa = (word, pairs, corpus_count, docs_count)
        insert_statement = "INSERT INTO irdb(word, file, frequency, count) VALUES(%s, %s, %s, %s)"
        try:
        # Executing the SQL command
            cursor.execute(insert_statement, dataa)
        
        # Commit your changes in the database
            conn.commit()

        except:
            not_added.append(word)
        # Rolling back in case of error
            conn.rollback()

        
        
            
        
conn.close()
```
