```python
url_check ={}
pno = 0
fail = 0
domain_names = {}
```

```python
try:
    while pno<300:
        #print(links)
        pno+=1
        cur_link = links.pop(0)
        #print(cur_link)
        try:
            page = requests.get(cur_link)
            #print(page.status_code)
        except:
            pno-=1
            fail+=1
            continue
        


        soup = BeautifulSoup(page.content, 'html.parser')
        page_text = ' '.join(soup.stripped_strings)
        content = lang_processing(page_text)
        
        title = (soup.find('title'))
        if title is None:
            title_csv = ''
        else:
            title_csv = lang_processing(title.get_text())
        #print(title.get_text())
        f =  open('pages/p'+str(pno)+'.txt', 'w')
        f.write(content)
        f.close()
        csv_row = [pno, cur_link,'p'+str(pno)+'.txt' , title_csv ]
        #print(csv_row)
        write_csv.writerow(csv_row)
        for link in soup.findAll('a', attrs={'href': re.compile("^https?://")}):
            add_link = link.get('href')
            add_link = getcleanurl(add_link)
            domain = getdomain(add_link)
            domain_names[domain] = True
            
            if url_check.get(add_link,False) or domain[-3:] != 'edu':
                continue
            else:
                links.append(link.get('href'))
                url_check[add_link] = True
                count_links+=1
        #print(links[-3:])
        #time.sleep(0.1)

    #csv_f.close()
except KeyboardInterrupt:
    pass
```

```python
#Language Processing - converts emoji to text, remove https links, removes stock ticker symbols, punctuation, perfoms lemmatization, etc

def lang_processing(text):
    #print("Cleaning Data (stop words, punctuation, emoji-conversion, lemmatize, )...")
    stop_words = set(stopwords.words("english"))

    keywords = []
        
    # remove URLs
    #print(re.findall(r'http\S+', text))
    text = re.sub(r'http\S+', '', text)   
    #text = emoji.demojize(text)
        
    # remove punctuations
    text = re.sub('[^a-zA-Z0-9]', ' ', text)
        
    # remove special characters
    text=re.sub("(\\W)+"," ",text)
        
    # convert to lowercase
    text = text.lower()
        
    #remove tags
    text=re.sub("&lt;/?.*?&gt;"," &lt;&gt; ",text)
        
    # Lemmatize
    #text = text.split()
    #lem = WordNetLemmatizer()
    #text = [lem.lemmatize(word) for word in text if not word in stop_words]
    #text = ' '.join(text)
        
     

    return text

#lang_processing(clean_text)
```
